```python
import re
import os
import math
from dataclasses import dataclass
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd
from tqdm import tqdm

# NLP
import nltk
from nltk.tokenize import sent_tokenize

import spacy

# Transformers sentiment
from transformers import pipeline

# -----------------------------
# Config
# -----------------------------
INPUT_XLSX = "Nitna_Reviews spreadsheet.xlsx"
OUT_CLEAN_CSV = "nitna_reviews_clean.csv"
OUT_ASPECT_SENTIMENT_CSV = "reviews_aspect_sentiment.csv"

# Download punkt if needed
try:
    nltk.data.find("tokenizers/punkt")
except LookupError:
    nltk.download("punkt")

# spaCy model
NLP = spacy.load("en_core_web_sm", disable=["ner", "parser"])
NLP.max_length = 2_000_000  # safety for long text


# -----------------------------
# Helpers: column inference
# -----------------------------
def _norm_col(c: str) -> str:
    return re.sub(r"[^a-z0-9]+", "_", str(c).strip().lower()).strip("_")

def infer_column(df: pd.DataFrame, candidates: List[str]) -> Optional[str]:
    """
    Return the first matching column based on normalized name containment.
    candidates: list of normalized tokens such as ["review_text", "text", "content"]
    """
    norm_map = {_norm_col(c): c for c in df.columns}
    norm_cols = list(norm_map.keys())

    for cand in candidates:
        # exact match
        if cand in norm_map:
            return norm_map[cand]
        # containment match
        for nc in norm_cols:
            if cand in nc:
                return norm_map[nc]
    return None


# -----------------------------
# Text cleaning
# -----------------------------
URL_RE = re.compile(r"(https?://\S+|www\.\S+)", re.IGNORECASE)
HTML_RE = re.compile(r"<.*?>")
MULTISPACE_RE = re.compile(r"\s+")

def clean_text(x: str) -> str:
    if pd.isna(x):
        return ""
    x = str(x)
    x = x.replace("\u200b", " ").replace("\ufeff", " ")
    x = HTML_RE.sub(" ", x)
    x = URL_RE.sub(" ", x)
    x = x.replace("&amp;", "&")
    x = re.sub(r"[\r\n\t]+", " ", x)
    x = MULTISPACE_RE.sub(" ", x).strip()
    return x

def simple_lemmatize(text: str) -> str:
    """
    Conservative lemmatization: keep meaning, avoid aggressive stemming.
    """
    if not text:
        return text
    doc = NLP(text)
    toks = []
    for t in doc:
        if t.is_space:
            continue
        # keep punctuation out (often not needed for aspect extraction)
        if t.is_punct:
            continue
        lemma = t.lemma_.lower().strip()
        if lemma:
            toks.append(lemma)
    return " ".join(toks)


# -----------------------------
# Aspect dictionary (editable)
# -----------------------------
ASPECT_KEYWORDS: Dict[str, List[str]] = {
    "absorbency": [
        "absorb", "absorbent", "absorbency", "leak", "leaks", "leaking", "leakage",
        "soak", "soaks", "soaking", "heavy flow", "overnight", "holds", "hold"
    ],
    "comfort": [
        "comfort", "comfortable", "soft", "softer", "itch", "itchy", "irritat", "rash",
        "chafe", "chafing", "gentle", "cozy"
    ],
    "price": [
        "price", "cost", "costly", "expensive", "cheap", "affordable", "value", "deal", "worth"
    ],
    "discretion": [
        "discreet", "discrete", "bulky", "bulk", "thin", "thick", "visible", "show",
        "noise", "noisy", "crinkle", "crinkly"
    ],
    "adhesive_fit": [
        "adhesive", "stick", "sticky", "stuck", "stays", "stay", "slip", "slips",
        "move", "moves", "shift", "shifts", "wings", "fit"
    ],
    "packaging": [
        "packaging", "pack", "wrapper", "wrap", "individually", "box", "bundle"
    ],
    "sustainability": [
        "sustain", "eco", "biodegrad", "organic", "cotton", "plastic", "compost",
        "environment", "green"
    ],
}

def sentence_mentions_aspect(sentence: str, aspect_terms: List[str]) -> bool:
    s = sentence.lower()
    # allow simple substring match; for multiword terms (e.g., "heavy flow") it's fine
    return any(term in s for term in aspect_terms)


# -----------------------------
# Transformer sentiment
# -----------------------------
@dataclass
class SentimentScore:
    label: str
    score: float  # model confidence/prob-like

def build_sentiment_pipeline(model_name: str = "distilbert-base-uncased-finetuned-sst-2-english"):
    """
    General sentiment model (POSITIVE/NEGATIVE). For ABSA, we run it on aspect-relevant sentences.
    You can swap to a more domain/ABSA-tuned model later.
    """
    return pipeline("sentiment-analysis", model=model_name, truncation=True)

def to_signed_score(label: str, score: float) -> float:
    """
    Convert POSITIVE/NEGATIVE + confidence to a signed score in [-1, 1].
    """
    label_u = label.upper()
    if "POS" in label_u:
        return float(score)
    if "NEG" in label_u:
        return -float(score)
    # if model returns other labels, treat as neutral-ish
    return 0.0

def aggregate_scores(scores: List[float]) -> float:
    """
    Aggregate sentence-level sentiment for an aspect within a review.
    """
    if not scores:
        return np.nan
    # robust mean (clip extremes)
    scores = np.clip(np.array(scores, dtype=float), -1, 1)
    return float(np.mean(scores))


# -----------------------------
# Main pipeline
# -----------------------------
def load_reviews(xlsx_path: str) -> pd.DataFrame:
    # If your file has multiple sheets, this reads the first by default
    df = pd.read_excel(xlsx_path)
    # Normalize column names (keep originals too)
    df.columns = [c.strip() if isinstance(c, str) else str(c) for c in df.columns]
    return df

def clean_reviews(df: pd.DataFrame) -> Tuple[pd.DataFrame, Dict[str, str]]:
    """
    Cleans dataset and returns (clean_df, colmap).
    colmap: resolved key fields.
    """
    df = df.copy()

    # Try to infer key columns
    text_col = infer_column(df, ["review_text", "review", "text", "content", "comment", "body"])
    brand_col = infer_column(df, ["brand", "brand_name", "product_brand"])
    rating_col = infer_column(df, ["rating", "stars", "star_rating", "score"])
    title_col = infer_column(df, ["title", "review_title", "headline"])
    date_col = infer_column(df, ["date", "review_date", "time"])

    if text_col is None:
        raise ValueError(
            "Could not infer the review text column. "
            "Rename your main text column to something like 'review_text' or 'review'."
        )

    # Create standardized columns
    df["review_text_raw"] = df[text_col]
    if title_col is not None:
        df["review_title_raw"] = df[title_col]
        df["review_text_raw"] = df["review_title_raw"].fillna("").astype(str) + ". " + df["review_text_raw"].fillna("").astype(str)

    df["review_text"] = df["review_text_raw"].apply(clean_text)
    # remove empty reviews
    df = df[df["review_text"].str.len() > 0].copy()

    # Optional: lemmatized version for keywording/topic modeling later
    df["review_text_lemma"] = df["review_text"].apply(simple_lemmatize)

    # Brand
    if brand_col is not None:
        df["brand"] = df[brand_col].astype(str).str.strip()
        df.loc[df["brand"].str.lower().isin(["nan", "none", ""]), "brand"] = np.nan
    else:
        df["brand"] = np.nan

    # Rating
    if rating_col is not None:
        df["rating"] = pd.to_numeric(df[rating_col], errors="coerce")
    else:
        df["rating"] = np.nan

    # Date
    if date_col is not None:
        df["review_date"] = pd.to_datetime(df[date_col], errors="coerce")
    else:
        df["review_date"] = pd.NaT

    # Dedupe (exact text duplicates)
    df = df.drop_duplicates(subset=["review_text"]).reset_index(drop=True)

    # Basic sanity trims
    df["review_text"] = df["review_text"].str.slice(0, 5000)  # keep it manageable
    df["review_text_lemma"] = df["review_text_lemma"].str.slice(0, 5000)

    colmap = {
        "text_col": text_col,
        "brand_col": brand_col or "",
        "rating_col": rating_col or "",
        "title_col": title_col or "",
        "date_col": date_col or "",
    }
    return df, colmap

def aspect_sentiment(df: pd.DataFrame, sent_pipe) -> pd.DataFrame:
    """
    For each review, find sentences mentioning each aspect; run sentiment on those sentences.
    Output: aspect sentiment scores per review (wide format).
    """
    rows = []
    for idx, text in tqdm(list(enumerate(df["review_text"].tolist())), desc="ABSA sentiment", total=len(df)):
        sents = sent_tokenize(text)
        aspect_scores = {}
        aspect_counts = {}

        for aspect, terms in ASPECT_KEYWORDS.items():
            matched = [s for s in sents if sentence_mentions_aspect(s, terms)]
            aspect_counts[f"{aspect}_n_sents"] = len(matched)

            if not matched:
                aspect_scores[aspect] = np.nan
                continue

            # sentiment per matched sentence
            signed_scores = []
            for s in matched:
                out = sent_pipe(s)[0]  # {'label': 'POSITIVE', 'score': ...}
                signed_scores.append(to_signed_score(out["label"], out["score"]))

            aspect_scores[aspect] = aggregate_scores(signed_scores)

        rows.append({"row_id": idx, **aspect_scores, **aspect_counts})

    out = pd.DataFrame(rows)
    return out

def main():
    df_raw = load_reviews(INPUT_XLSX)
    df_clean, colmap = clean_reviews(df_raw)

    print("Inferred column mapping:", colmap)
    print("Clean rows:", len(df_clean))
    df_clean.to_csv(OUT_CLEAN_CSV, index=False)

    # Sentiment model (can swap model_name)
    sent_pipe = build_sentiment_pipeline(
        model_name="distilbert-base-uncased-finetuned-sst-2-english"
    )

    df_aspect = aspect_sentiment(df_clean, sent_pipe)

    # Merge back to cleaned reviews
    df_out = pd.concat([df_clean.reset_index(drop=True), df_aspect.drop(columns=["row_id"])], axis=1)
    df_out.to_csv(OUT_ASPECT_SENTIMENT_CSV, index=False)

    print("Saved cleaned data to:", OUT_CLEAN_CSV)
    print("Saved aspect sentiment data to:", OUT_ASPECT_SENTIMENT_CSV)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import re

xlsx_path = "Nitna_Reviews spreadsheet.xlsx"   # exactly as shown in your folder
out_clean = "nitna_reviews_clean_FIXED.csv"

df_raw = pd.read_excel(xlsx_path)

# Clean text
def clean_text(x):
    if pd.isna(x): 
        return ""
    x = str(x)
    x = re.sub(r"<.*?>", " ", x)                       # HTML
    x = re.sub(r"(https?://\S+|www\.\S+)", " ", x)     # URLs
    x = re.sub(r"[\r\n\t]+", " ", x)
    x = re.sub(r"\s+", " ", x).strip()
    return x

# Parse rating from "4.0 out of 5 stars"
def parse_rating(x):
    if pd.isna(x):
        return np.nan
    s = str(x)
    m = re.search(r"(\d+(\.\d+)?)", s)
    return float(m.group(1)) if m else np.nan

# Brand from product name (first token)
def parse_brand(product_name):
    if pd.isna(product_name):
        return np.nan
    s = str(product_name).strip()
    if not s or s.lower() in ["nan", "none"]:
        return np.nan
    return s.split()[0].strip(",").strip()

# Build clean df using your known columns
df = pd.DataFrame()
df["review_text_raw"] = df_raw["Review Text"]
df["review_text"] = df["review_text_raw"].apply(clean_text)

df["rating_raw"] = df_raw["Rating (Max 5)"]
df["rating"] = df["rating_raw"].apply(parse_rating)

df["review_date_raw"] = df_raw["Review Date"]
df["review_date"] = pd.to_datetime(df["review_date_raw"], errors="coerce", dayfirst=True)

df["product_name"] = df_raw["Product Name"].astype(str).str.strip()
df["brand"] = df["product_name"].apply(parse_brand)

# Drop empty text rows
df = df[df["review_text"].str.len() > 0].copy()

# Drop duplicates
df = df.drop_duplicates(subset=["product_name", "review_text"]).reset_index(drop=True)

df.to_csv(out_clean, index=False)
print("Saved:", out_clean)
print(df[["rating_raw", "rating", "brand"]].head(10))
print("Missing rating:", df["rating"].isna().sum(), "of", len(df))
print("Missing brand:", df["brand"].isna().sum(), "of", len(df))
```

```python
import pandas as pd
import numpy as np

# ✅ USE THE MERGED MODELING DATASET
DATA_PATH = "nitna_modeling_dataset.csv"

df = pd.read_csv(DATA_PATH)

# --- Basic sanity ---
print("Rows:", len(df))
print("Columns:", len(df.columns))
print(df[["rating", "brand"]].head())

# --- Define prediction targets ---
df["loyal"] = (df["rating"] >= 4).astype(int)

# --- Choose model features ---
exclude_cols = {
    "rating", "rating_raw", "Review Text", "review_text",
    "review_date", "product_name", "brand", "loyal"
}

sent_cols = [
    c for c in df.columns
    if c not in exclude_cols
    and not c.endswith("_n_sents")
    and df[c].dtype in ["float64", "int64"]
]

count_cols = [c for c in df.columns if c.endswith("_n_sents")]

print("Sentiment columns:", sent_cols)
print("Count columns:", count_cols)

feature_cols = sent_cols + count_cols

use_brand = (
    "brand" in df.columns
    and df["brand"].notna().sum() > 0
    and df["brand"].nunique() > 1
)

if use_brand:
    feature_cols.append("brand")

print("Using brand as feature?", use_brand)
print("Num sentiment features:", len(sent_cols))
print("Num count features:", len(count_cols))
print("Total features:", len(feature_cols))

# X, y
X = df[feature_cols].copy()
y_class = df["loyal"].copy()
y_reg = df["rating"].copy()

# Fill missing values
for c in sent_cols + count_cols:
    X[c] = X[c].fillna(0)

if use_brand:
    X["brand"] = X["brand"].fillna("Unknown").astype(str)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report

# Targets
df["loyal"] = (df["rating"] >= 4).astype(int)
y = df["loyal"]

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

numeric_features = [c for c in feature_cols if c != "brand"]
categorical_features = ["brand"] if use_brand else []

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
    ],
    remainder="drop"
)

# Models
logreg = Pipeline([
    ("prep", preprocess),
    ("model", LogisticRegression(max_iter=3000, class_weight="balanced"))
])

rf = Pipeline([
    ("prep", preprocess),
    ("model", RandomForestClassifier(
        n_estimators=600,
        random_state=42,
        class_weight="balanced_subsample"
    ))
])

# Fit
logreg.fit(X_train, y_train)
rf.fit(X_train, y_train)

def eval_model(name, model):
    proba = model.predict_proba(X_test)[:, 1]
    pred = (proba >= 0.5).astype(int)
    auc = roc_auc_score(y_test, proba)
    print(f"\n=== {name} ===")
    print("ROC AUC:", round(auc, 4))
    print(classification_report(y_test, pred, digits=3))
    return auc, proba, pred

auc_lr, proba_lr, pred_lr = eval_model("Logistic Regression", logreg)
auc_rf, proba_rf, pred_rf = eval_model("Random Forest", rf)

best_name, best_model, best_proba, best_pred = (
    ("Random Forest", rf, proba_rf, pred_rf) if auc_rf >= auc_lr else ("Logistic Regression", logreg, proba_lr, pred_lr)
)

print("\nBest model:", best_name)
```
